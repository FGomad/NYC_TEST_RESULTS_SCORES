Every year, American high school students take SATs, which are standardized tests intended to measure literacy, numeracy, and writing skills. There are three sections - reading, math, and writing, each with a **maximum score of 800 points**. These tests are extremely important for students and colleges, as they play a pivotal role in the admissions process.

Analyzing the performance of schools is important for a variety of stakeholders, including policy and education professionals, researchers, government, and even parents considering which school their children should attend.

**Best Maths Schools -  That is Schools with at least 80 %, of the Possible 800 score for Maths**

```python
#Creating a subset from the main dataframe which contains school name and average maths

maths_performance_in_schools = schools [['school_name', 'average_math']]

maths_performance_in_schools.head()

#Filter the maths_performance_in_schools with respect to schools with schools which had at least 80% of the totall scores for maths

best_math_schools_unsorted = maths_performance_in_schools[maths_performance_in_schools['average_math'] >= 640] # 80% of possible 800 is 640

best_math_schools = best_math_schools_unsorted.sort_values('average_math', ascending =False)

print(best_math_schools)
```
